## OUTLIERS

### DETECTING OUTLIERS

```python
data1.boxplot(figsize = [24,8])
```

```python
def out_by_iqr(data):
    out = []
    data = np.sort(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lwr_bnd = q1 - (1.5 * iqr)
    upr_bnd = q3 + (1.5 * iqr)
    for i in data:
        if (i < lwr_bnd or i > upr_bnd):
            out.append(i)
    return out    
```

### HANDLING OUTLIERS

```python
def out_iqr(data, col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    lwr = q1 - 1.5 * iqr
    upr = q3 + 1.5 * iqr
    return lwr, upr 
```

```python
for i in data.columns:
    lwr, upr = out_iqr(data1, i)
    data[i].clip(lwr, upr, inplace = True)
data
```
